```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Set the random seed for reproducibility
np.random.seed(42)
```

# Generating Data

```python
X = np.arange(1,50)
y = 3 * X + 5 
X,y
```

```python
X = X.reshape(-1,1)
```

```python
y = y.reshape(-1,1)
```

```python
# Plot the generated data
plt.scatter(X, y, color='blue')
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.title("Generated Data")
plt.show()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.2)

# Plot the training and testing data
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='red', label='Testing Data')
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.title("Training and Testing Data")
plt.legend()
plt.show()
```

# Training a model

```python
y_pred_untrained = np.random.randint(140, size=(len(y_test)))

plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_pred_untrained, color='red', label='Untrained Data')
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model = model.fit(X_train, y_train)
y_pred = model.predict(X_test)

plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_pred_untrained, color='red', label='Untrained Data')
plt.scatter(X_test, y_pred, color='orange', label='Trained Data')
plt.legend()
```

```python
model.coef_[0][0], model.intercept_[0]
```

# Predicting new data

```python
model.predict([[5.5]])
```

# Random Data

```python
X = 2 * np.random.rand(100, 1)  # 100 data points, one feature
y = 3 * X + 5 + np.random.randn(100, 1)  # Linear relationship with some noise

plt.scatter(X, y, color='blue')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

# Extract the model's parameters
intercept = model.intercept_[0]
coefficient = model.coef_[0][0]

X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_line = intercept + coefficient * X_range
```

```python
# Plotting the line
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='red', label='Test Data') 

plt.plot(X_range, y_line, color='green', label='Fitted Line')


# Plotting predicitions
y_pred = model.predict(X_test)
plt.scatter(X_test, y_pred, color='orange', label='Test Data')
```
